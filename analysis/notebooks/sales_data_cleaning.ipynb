```python
import pandas as pd
```

```python
sales_df = pd.read_excel("../../data/raw_US_Regional_Sales_Data.xlsx", sheet_name=None)
```

```python
sales_sheet = sales_df["Sales Orders Sheet"]
```

```python
sales_sheet.head(5)
```

```python
customer_sheet = sales_df["Customers Sheet"]
```

```python
customer_sheet.set_index(customer_sheet.columns[0], inplace=True)
```

```python
customer_sheet.head(5)
```

```python
merged_df = pd.merge(sales_sheet, customer_sheet, on='_CustomerID', how='left')
```

```python
merged_df.head(5)
```

```python
products_sheet = sales_df["Products Sheet"]
products_sheet.set_index(products_sheet.columns[0], inplace=True)
products_sheet.head(5)
```

```python
merged_df = pd.merge(merged_df, products_sheet, on='_ProductID', how='left')
merged_df.head(5)
```

```python
team_sheet = sales_df["Sales Team Sheet"]
sub_team_sheet = team_sheet[["_SalesTeamID", "Sales Team"]]
sub_team_sheet.set_index(sub_team_sheet.columns[0], inplace=True)
sub_team_sheet.head(5)
```

```python
merged_df = pd.merge(merged_df, sub_team_sheet, on="_SalesTeamID", how="left")
merged_df.head(5)
```

```python
store_sheet = sales_df["Store Locations Sheet"]
sub_store_sheet = store_sheet[["_StoreID", "City Name", "StateCode", "Population", "Household Income", "Median Income"]]
sub_store_sheet.set_index(sub_store_sheet.columns[0], inplace=True)
sub_store_sheet.head(5)
```

```python
merged_df = pd.merge(merged_df, sub_store_sheet, on="_StoreID", how="left")
merged_df.head(5)
```

```python
region_sheet = sales_df["Regions Sheet"]
sub_region_sheet = region_sheet[["StateCode", "State"]]
sub_region_sheet.set_index(sub_region_sheet.columns[0], inplace=True)
sub_region_sheet.sample(5)
```

```python
merged_df = pd.merge(merged_df, sub_region_sheet, on="StateCode", how="left")
merged_df.sample(5)
```

```python
print(merged_df.columns)
```

```python
merged_df.drop(columns=["_SalesTeamID", "_CustomerID", "_StoreID", "_ProductID", "StateCode"], inplace=True)
```

```python
merged_df.sample(5)
```

```python
print(merged_df.isna().sum())
```

```python
merged_df["Unit Profit"] = merged_df["Unit Price"] - merged_df["Unit Cost"]
merged_df.head(5)
```

```python
merged_df.to_excel("../../data/cleaned_US_Regional_Sales_Data.xlsx", index=False)
```
